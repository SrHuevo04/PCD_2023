### Hilos.vs.Procesos

#### Ejemplos de hilos

```python
import threading as th
```

```python
def doNothing():
    while True:
        pass
```

```python
for i in range(20):
    th.Thread(target=doNothing).start()

th.enumerate()
```

#### Ejemplo procesos

```python
for i in range(20):
    mp.Process(target = doNothing).start
```
